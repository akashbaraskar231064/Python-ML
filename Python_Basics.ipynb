```python
numbers = []

print("Enter numbers one by one. Type 'done' to finish.")

while True:
    user_input = input("Enter a number: ")

    # Check if the user wants to stop
    if user_input.lower() == 'done':
        break

    try:
        # Convert input to a float to handle decimals
        num = float(user_input)
        numbers.append(num)
    except ValueError:
        print("Invalid input. Please enter a number or 'done'.")

# Check if the list isn't empty before finding max
if numbers:
    max_val = max(numbers)
    print(f"The maximum number entered is: {max_val}")
else:
    print("No numbers were entered.")
```

```python
class Car:
  def __init__(self, brand, model, year):
    self.brand = brand
    self.model = model
    self.year = year

  def display_info(self):
    print(f"Car: {self.brand} {self.model} {self.year}")

car1 = Car("Toyota", "Corolla", 2022)
car2 = Car("Tesla", "Model S", 2023)

car1.display_info()
car2.display_info()
```

```python
class Employee:
  def __init__(self, Name, Id,  Salary):
    self.Name = Name
    self.Id = Id
    self.Salary = Salary

  def display_info(self):
    print(f'Employee: {self.Name} (self.Id) {self.Salary}')

emp1 = Employee("Avi", "a1", "55000")
emp2 = Employee("Dipesh", "d1", "50000")

emp1.display_info()
emp2.display_info()
```

```python
class Student:
  def __init__(self, Name, Marks,  Course):
    self.Name = Name
    self.Marks = Marks
    self.Course = Course

  def display_info(self):
    print(f'Student: {self.Name} {self.Marks} {self.Course}')

stu1 = Employee("Avi", "55", "B.tech")
stu2 = Employee("Dipesh", "45", "B.tech")

stu1.display_info()
stu2.display_info()
```
